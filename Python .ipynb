```python
def count_frequency(elements):
    frequency = {}
    for x in elements:
        if x in frequency:
            frequency[x] += 1
        else:
            frequency[x] = 1
    return frequency  # Move this outside the loop

data = [1, 2, 2, 3, 3, 3, 4]
frequencies = count_frequency(data)
print(frequencies)
```

```python
# Merge two dictionaries and sum the values of common keys
def merge_and_sum(dict1, dict2):
    merged = dict1.copy()  # Start with a copy of the first dictionary
    for key, value in dict2.items():
        if key in merged:
            merged[key] += value  # Sum values for common keys
        else:
            merged[key] = value  # Add new keys
    return merged

# Example usage
dict1 = {"a": 10, "b": 20, "c": 30}
dict2 = {"b": 15, "c": 25, "d": 40}
result = merge_and_sum(dict1, dict2)
print(result)
```
