The try and except blocks in Python are used for handling exceptionsâ€”errors that occur during the execution of a program. This feature allows you to gracefully manage errors without stopping the program abruptly.

# Type of Errors

**ZeroDivisionError**

```python
try:
  a = int(input('Enter the first number:'))
  b = int(input('Enter the second number:'))
  c = a/b
  print(c)

except:
   print("enter the valid number")
```

**Index Error**

```python
try:
  index = [1,2,3,4,5]
  print(index[8])
except:
   print("index is out of range")
```

**Key Error**

```python
try:
  dict1 = {a:'1',b:'2'}
  print(dict1['c'])
except:
   print("it's a key error")
```

**Type Error**

```python
try:
  print("2" + 3)

except:
   print("Type error")
```

**Value Error**

```python
try:
  a = int(input('Enter the first number:'))
  b = int(input('Enter the second number:'))
  c = a/b
  print(c)

except:
   print("Value Error")
```

# Handling Specific Exception




```python
try:
  a = int(input('Enter the first number:'))
  b = int(input('Enter the second number:'))
  c = a/b
  print(c)

except ZeroDivisionError:
   print("Zero Division Error")

except ValueError:
   print("Value Error")

```

* An exception in Python is an event that occurs during the execution of a program and disrupts its normal flow. Exceptions are Python's way of signaling that an error or an unexpected condition has been encountered

# finally

```python
# it work's both error and non-error

try:
  a = int(input())
  b = int(input())
  c= a/b
  print(c)
finally:
  print('end')
```

# Using While Loop

```python
while True:
  try:
    a = int(input())
    b = int(input())
    c= a/b
  except:
    print("enter valid number")
  else:
    print("\nResult:",c)
    break
```

```python
while True:
  try:
    a = int(input())
    b = int(input())
    c= a/b

  except ZeroDivisionError:
    print("Zero Division Error")

  except ValueError:
    print("Value Error")
  else:
    print("\nResult:",c)
    break
```

```python
try:
   # Outer try block to handle user input
   a = int(input("Enter the number:"))
   b = int(input("Enter the number:"))

   try:
      # Inner try block to handle division
      c = a/b
      print(f"result:{c}")
   except ZeroDivisionError:
      print("-Cannot divide by zero")
   finally:
      print("-Inner Block Finished")

except ValueError:
   print("-Invalid input:please enter integer only.")
finally:
   print("-Outer block Finished")
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
def draw_h_pattern():
    # Top Pyramid
    for i in range(1, 6):
        spaces = ' ' * (5 - i)
        stars = 'H' * (2 * i - 1)
        print(spaces + stars)

    # Two vertical blocks of H
    for _ in range(7):
        print('  HHHHH               HHHHH')

    # Middle wide block
    for _ in range(3):
        print('  HHHHHHHHHHHHHHHHHHHHHHHHHHHHH')

    # Two vertical blocks of H again
    for _ in range(7):
        print('  HHHHH               HHHHH')

    # Bottom inverted pyramid
    for i in range(5, 0, -1):
        spaces = ' ' * (5 - i)
        stars = '    H    ' * (2 * i - 1)
        print(spaces + stars)

# Call the function
draw_h_pattern()
```

```python
def print_reverse_pyramid():
    # Loop to print the inverted pyramid
    for i in range(5, 0, -1):  # Start from 5 and go down to 1
        spaces = ' ' * (5 - i)  # Number of spaces decreases as i increases
        stars = 'H' * (2 * i - 1)  # The number of 'H' increases
        print(spaces + stars)

# Call the function
print_reverse_pyramid()
```
