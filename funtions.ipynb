```python
##print even numbers btw 1 and 20
def is_even(n):
  for i in range(1,n+1):
    if i%2==0:
      print(i)
is_even(20)
```

```python
#user input and count the number of vowels
def count_vowels():
  a=str(input("enter a word:"))
  b="AEIOUaeiou"
  count=0
  for x in a:
   if x in b:
    count=count+1
  print(count)
count_vowels()
```

```python
#add first 20 natural numbers
def add_sum(n):
  a=0
  for i in range(1,n+1):
    a+=i
  print(a)
add_sum(20)
```

```python
#multiply first 50 natural numbers
def add_multiply(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  print(a)
add_multiply(50)
```
