# PYTHON if...elif...else method

```python
                                               #python if....elif....else method
a=3
b=8
if a>b:
  print("a is greater than b")
elif a==b:
  print("a is equal to b")
else:
  print("a is less than b")
```

```python
#get a user input value,and check 'if' it is "positive","negative or "zero"
a=int(input("enter a number:"))
if a>0:
  print("positive")
elif a==0:
  print("zero")
else:
  print("negative")
```

```python
#check if any element is multiple of 5 and positive.else print 'else' print 'not in multiple of 5'
a=[25,32,45,77,0,-5,-20]
for x in a:
  if x%5==0 and x>0:
   print(f"{x}:five is positive")
else:
   print(f"{x}:not in multiple of 5")
```

# while loop

```python
                                              #PYTHON WHILE LOOP
#print 0-10
i=0
while i<=10:
  print(i)
  i=i+1
```

```python
#print 10-0
i=10
while i>=0:
  print(i)
  i=i-1
```

```python
i=1
while i<=10:                      #break statement
  print(i)
  if i==5:
    break
  i=i+1
```

```python
                                      #continue statement
i=1
while i<=10:
  i=i+1
  if i==5:
    continue
  print(i)
```

# for loop

```python
                                                 #FOR LOOP
x="hello world"
for i in x:
  print(i,end="    ")
```

```python
a=["apple","orange","mango"]
for i in a:                             #break method in for loop
  if i=="orange":
    break
  print(i)
```
